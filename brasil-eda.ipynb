### Visto todo lo anterior, me dispongo a unir el dataframe 

```python
#https://stackoverflow.com/questions/11479392/what-does-a-for-loop-within-a-list-do-in-python  buscando como hacer el loop mas eficiente

year = range(2013, 2022)

lista_df = [pd.read_csv(f"datos/datos-{y}.csv", sep=";", encoding='latin-1') for y in year]

df_all = pd.concat(lista_df, axis=0)  ## concat admite una lista de dataframes

df_all.columns
```

```python
## https://www.statology.org/pandas-rename-columns-with-dictionary/

## Renombrar columnas del dataframe df_all

column_translation_dict = {
 'CÓDIGO ÓRGÃO SUPERIOR': 'codigo_organo_superior',
 'NOME ÓRGÃO SUPERIOR': 'nombre_organo_superior',
 'CÓDIGO ÓRGÃO': 'codigo_organo',
 'NOME ÓRGÃO': 'nombre_organo',
 'CÓDIGO UNIDADE GESTORA': 'codigo_unidad_gestora',
 'NOME UNIDADE GESTORA': 'nombre_unidad_gestora',
 'CATEGORIA ECONÔMICA': 'categoria_economica',
 'ORIGEM RECEITA': 'origen_receta',
 'ESPÉCIE RECEITA': 'especie_receta',
 'DETALHAMENTO': 'detallamiento',
 'VALOR PREVISTO ATUALIZADO': 'valor_previsto_actualizado',
 'VALOR LANÇADO': 'valor_lanzado',
 'VALOR REALIZADO': 'valor_realizado',
 'PERCENTUAL REALIZADO': 'percentual_realizado',
 'DATA LANÇAMENTO': 'fecha_lanzamiento',
 'ANO EXERCÍCIO': 'ano_ejercicio'
}



#define dictionary
#rename columns in DataFrame using dictionary

df_all.rename(columns=column_translation_dict, inplace=True)

df_all.columns

df_all.info()
```

## Revisar las columnas, tipos de datos

```python
#df_all.isna().sum()

df_all.info()

#null_mask = df_all.isnull().any(axis=1)
#null_rows = df_all[null_mask]

#print(null_rows)
## https://stackoverflow.com/questions/8420143/valueerror-could-not-convert-string-to-float-id

df_all["valor_previsto_actualizado"] = [float(str(i).replace(",", ".")) for i in df_all["valor_previsto_actualizado"]]
df_all['valor_previsto_actualizado'] = df_all['valor_previsto_actualizado'].astype('float')


df_all["valor_realizado"] = [float(str(i).replace(",", ".")) for i in df_all["valor_realizado"]]
df_all['valor_realizado'] = df_all['valor_realizado'].astype('float')

df_all["valor_lanzado"] = [float(str(i).replace(",", ".")) for i in df_all["valor_lanzado"]]
df_all['valor_lanzado'] = df_all['valor_lanzado'].astype('float')


df_all["percentual_realizado"] = [float(str(i).replace(",", ".")) for i in df_all["percentual_realizado"]]
df_all['percentual_realizado'] = df_all['percentual_realizado'].astype('float')

df_all['ano_ejercicio'] = df_all['ano_ejercicio'].astype('int')

df_all['fecha_lanzamiento'] = pd.to_datetime(df_all['fecha_lanzamiento'], format="%d/%m/%Y")

#df_porcentaje_uniques = df_all['PORCENTAJE REALIZADO'].unique()
#df_porcentaje_uniques

df_all.info()
```
